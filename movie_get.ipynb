# Getting the top movies of each summer

We will scrape BoxOfficeMojo for the top N movies of the summer (first week in May - Labor Day weekend release dates) for the past ten years.

```python
import pickle
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
prefix = 'http://www.boxofficemojo.com'
suffixes = ['?chart=&yr=2015&season=Summer']
```

```python
page1 = requests.get(prefix + '/seasonal/' + suffixes[0])
page1.status_code
```

```python
page1text = page1.text
soup1 = bs(page1text)
print soup1.prettify()
```

```python
for suffix in soup1.find_all('a', href = re.compile("^\?chart"))[3:12]:
    suffixes.append(suffix.get('href'))
suffixes
    
```

```python
pages = []
for suffix in suffixes:
    pages.append(requests.get(prefix + '/seasonal/' + suffix))
pages
```

```python
soups = []
for page in pages:
    soups.append(bs(page.text))
print soups[9].prettify()
```

```python
with open('top_summer_movie_pages.pkl', 'w') as f:
    pickle.dump(pages, f)
```

```python
soups = []
with open('top_summer_movie_pages.pkl', 'r') as f:
    pages = pickle.load(f)
soups = map(lambda x: bs(x.text), pages)
print soups[0].find_all('table')[3].prettify()
```

```python
print soups[0].find_all('table')[3].find_all('tr')[2].prettify(), len(soups)
```

```python
def movie_strip_dollar(s):
    if s[0] == '$':
        return float(s[1:].replace(',', ''))
    return s

top500 = []
for i in range(len(soups)):
    top_table = soups[i].find_all('table')[3]
    for line in top_table.find_all('tr')[2:52]:
        row = []
        for cell in line.find_all('td'):
            for string in cell.strings:
                row.append(movie_strip_dollar(str(string)))
        row.append(2015 - i)
        top500.append(row)
pd.DataFrame(top500)
```

```python
with open('top500movies.pkl', 'w') as f:
    pickle.dump(top500, f)
```

```python
with open('top500movies.pkl', 'r') as f:
    new_df = pickle.load(f)
```

```python
new_df = pd.DataFrame(new_df, columns = ['index', 'title', 'prod', 'gross', 'trs', 'open', 'open_trs', 'op_date', 'cl_date', 'year'])
```
