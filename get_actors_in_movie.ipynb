# A first pass at pulling actors from a movie page

Here, we will start by pulling the actors from Jurassic World.

```python
import pickle
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
url = 'http://www.imdb.com/search/title?boxoffice_gross_us=2900000,&count=250&release_date=2006-05-01,2015-09-30&sort=boxoffice_gross_us,asc&title_type=feature,documentary&view=simple'
```

```python
page = requests.get(url)
```

```python
pages[0] = page.text
```

```python

soups = [bs(page.text)]
suffix = '/search/title?boxoffice_gross_us=2900000,&count=250&release_date=2006-05-01,2015-09-30&sort=boxoffice_gross_us,asc&title_type=feature,documentary&view=simple'
i = 0
```

```python
soups[0].find(text = re.compile("Next")).parent.get('href')
```

```python
len(soups)
```

```python
while suffix:
    link = soups[i].find(text = re.compile(u'Next\xa0\xbb'))
    if link:
        suffix = link.parent.get('href')
        page = requests.get('http://www.imdb.com' + suffix).text
        pages.append(page)
        soups.append(bs(page))
        i += 1
    else:
        suffix = None
print len(soups)
```

```python
with open('imdb_summer_movie_links.pkl', 'w') as imdb_links:
    pickle.dump(pages, imdb_links)
```

```python
def get_imdb_link(movie, i):
    for soup in soups:
        spot = soup.find(text = re.compile(movie.lower(), re.I))
        if spot:
            return spot.parent.get('href')
    print i
    return
        

def get_imdb_links(movies):
    links = []
    i = 0
    count = 0
    for movie in movies:
        link = get_imdb_link(movie, i)
        links.append(link)
        i += 1
        if link == None:
            count += 1
    print count
    return links
```

```python
links_take_1 = get_imdb_links(a.values)
with open("imdb_movie_links.pkl", 'w') as imdb_movie_links:
    pickle.dump(links_take_1, imdb_movie_links)
```
